```python
from collections import defaultdict

def convert_file(input_path, output_path):
    with open(input_path, "r") as f:
        content = f.read()

    nums = list(map(int, content.strip().split(",")))
    counts = defaultdict(int)
    result = []

    for num in nums:
        tag = counts[num]
        float_val = num + tag * 0.001
        result.append(f"{float_val:.3f}")
        counts[num] += 1

        if (counts[num] > 1):
            print(num)

    with open(output_path, "w") as f:
        f.write(", ".join(result))


if __name__ == "__main__":
    convert_file("./input_data/stability.txt", "./input_data/stability.txt")
```
